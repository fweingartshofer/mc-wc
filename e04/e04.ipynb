```python
import pandas as pd

folder = 'data'

patients = pd.read_csv(f'{folder}/patient-info.csv')

file_path = f'{folder}/{str(patients.iloc[0]["File"]).zfill(3)}.txt'
file_path
```

```python
import heartpy as hp
for patient in patients.values:
    try:
        file_path = f'{folder}/{str(patient[0]).zfill(3)}.txt'
        series = pd.read_csv(file_path, header=None).squeeze()
        timestamps = pd.TimedeltaIndex([x*7.8125 for x in range(0, len(series))])
        series = series.set_axis(timestamps)
        rr = hp.process(series, sample_rate=128)
        print(patient[0], ".txt " , patient[1], "years -", rr[1]['breathingrate'] * 60)
    except Exception as e:
        pass
        #print("Error while calculating breathing rate", e)
```

```python
file_path = f'{folder}/{str(patients.iloc[0]["File"]).zfill(3)}.txt'
series = pd.read_csv(file_path, header=None).squeeze()
timestamps = pd.TimedeltaIndex([x*7.8125 for x in range(0, len(series))])
series = series.set_axis(timestamps)
```

```python
series.plot(figsize=(25, 10))
```

```python
offset_series = pd.concat([pd.Series([0]), series])
for i in range(0, len(series)):
    x = offset_series.iloc[i]
#    y = series.iloc[i]
#    series.iloc[i] = y - x
series.plot(figsize=(25,10))
```

```python
from scipy import signal
resampled = signal.resample(series, 200)
pd.DataFrame(resampled).plot(figsize=(25, 10))
```

```python


X = pd.Series(resampled)
X.plot(figsize=(25, 10))
```

```python

import numpy as np
rr_intervals = np.sort(X)
rr_diff = np.diff(rr_intervals)

# Calculate the average of the differences
breathing_rate = np.mean(rr_diff)

print('Breathing rate:', breathing_rate, 'breaths per minute')
```
